```python
import pandas as pd
import joblib as jb

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/home/orange/EduML/data/external/student_performance.csv')

features = df.columns.drop(['StudentID', 'GPA'])
X = df[features]
y = df['GPA']

X = pd.get_dummies(X, 
                   columns=['Ethnicity', 'ParentalEducation', 'GradeClass', 'Music', 'Extracurricular', 'Sports'], 
                   drop_first=True)
```

```python
lin_model = jb.load('/home/orange/EduML/models/gpa_target/EduML-lin.pkl')
dt_model = jb.load('/home/orange/EduML/models/gpa_target/EduML-dt.pkl')
xgb_model = jb.load('/home/orange/EduML/models/gpa_target/EduML-xgb.pkl')

lin_y = lin_model.predict(X)
dt_y = dt_model.predict(X)
xgb_y = xgb_model.predict(X)
```

```python
# linear regression
sns.kdeplot(lin_y, fill=True)
plt.xlabel(' GPA')
plt.ylabel('')
plt.title('   GPA (lin regression)')
plt.show()

# xgboost
sns.kdeplot(xgb_y, fill=True)
plt.xlabel(' GPA')
plt.ylabel('')
plt.title('   GPA (xgboost)')
plt.show()

# decision trees
sns.kdeplot(dt_y, fill=True)
plt.xlabel(' GPA')
plt.ylabel('')
plt.title('   GPA (decision trees)')
plt.show()
```
